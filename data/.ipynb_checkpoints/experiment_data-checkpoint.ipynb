```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
conn = sqlite3.connect("data.db")
act = pd.read_sql_query("select * from actors;", conn)
act[['tothitsexp0', 'tothitsexp1', 'tothitsexp2', 'tothits']]
n_hits = np.array(act[['tothitsexp0', 'tothitsexp1', 'tothitsexp2']])
```

```python
def exp_format(x, pos=None):
    names = {1: 'Delay',
            2: 'Delay + predictor',
            3: 'No delay'}
    return names[x]
```

```python
fig1, ax1 = plt.subplots(dpi=100)
ax1.set_title('Basic Plot')
ax1.boxplot(n_hits)
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(exp_format))
plt.title('Number of hits in 90 seconds')
# plt.xlabel('Experiment')
plt.ylabel('Number of hits')
plt.show()
```
